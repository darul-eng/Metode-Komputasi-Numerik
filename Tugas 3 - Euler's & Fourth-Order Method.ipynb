```python
# library akar kuadrat dari math
from math import sqrt
# dictionary
data = {}
data['Xi']=[]
data['Y']=[]
# pilih jenis metode yang di inginkan
print ("Pilih Jenis order dari runge-kutta")
print ("1. first/Euler's")
print ("2. fourth")
while True:
    pilihan = input("Masukkan pilihan(1 atau 2) = ")
    #jika yang di pilih first
    if pilihan == '1':
        print ("First-order runge-kutta/Eulers method")
        def fungsi(x,y):
            return (1+(2*x))*sqrt(y)
        #input nilai h
        while True:
            while True:
                #masukkan nilai h
                try:
                    h = float(input("Masukkan nilai h = "))
                    break
                except ValueError:
                    print("Silakan masukkan angka yang valid")
                    
            while True :
                #Masukkan x awal         
                try:
                    x_awal = float(input("Masukkan x awal = "))
                    break
                except ValueError:
                    print("Silakan masukkan angka yang valid.\n")

            while True:
                #Masukkan x akhir         
                try:
                    x_akhir = float(input("Masukkan x akhir: "))

                    # masukkan ulang x akhir jika x akhir < x awal
                    if x_akhir > x_awal:
                        break
                    else:
                        print("Silakan masukkan angka yang lebih besar dari x awal\n")
                except ValueError:
                    print("Silakan masukkan angka yang valid.\n")
            if x_akhir > x_awal:
                break
        print("x awal = ",x_awal)
        print("x akhir = ",x_akhir)
        #input nilai y_awal         
        y_awal = float(input("Masukkan nilai y awal = "))
        print("Nilai y awal = ",y_awal)
        print("____________________________________")
        data['Xi'].append(x_awal)
        data['Y'].append(y_awal)
        while True:
            print ("Nilai x =",round((x_awal+h),3))
            fungsi_next = y_awal+(fungsi(x_awal,y_awal)*h)
            print("Nilai y(",round((x_awal+h),3),") = ",fungsi_next)
            x_awal+=h
            y_awal = fungsi_next
            data['Xi'].append(x_awal)
            data['Y'].append(y_awal)
            print("- - - - - - - - - - - - - - -- - - -")
            if x_awal >= x_akhir:
                break
        break
    elif pilihan == '2':
        print("fourth-order runge-kutta method")
        def fungsi(x,y):
            return (1+(2*x))*sqrt(y)
        def k_1(x,y):
            return fungsi(x,y)
        def k_2(x,y,h,k1):
            return fungsi(x+(0.5*h),y+(0.5*(k_1(x,y))*h))
        def k_3(x,y,h,k1,k2):
            return fungsi(x+(0.5*h),(y+(0.5*k2*h)))
        def k_4(x,y,h,k1,k2,k3):
            return fungsi((x+h),(y+(k3*h)))
        while True:
            while True:
                #masukkan nilai h
                try:
                    h = float(input("Masukkan nilai h = "))
                    break
                except ValueError:
                    print("Silakan masukkan angka yang valid")
                    
            while True :
                #Masukkan x awal         
                try:
                    x_awal = float(input("Masukkan x awal = "))
                    break
                except ValueError:
                    print("Silakan masukkan angka yang valid.\n")

            while True:
                #Masukkan x akhir         
                try:
                    x_akhir = float(input("Masukkan x akhir: "))

                    # masukkan ulang x akhir jika x akhir < x awal
                    if x_akhir > x_awal:
                        break
                    else:
                        print("Silakan masukkan angka yang lebih besar dari x_l\n")
                except ValueError:
                    print("Silakan masukkan angka yang valid.\n")
            if x_akhir > x_awal:
                break
        print("x awal = ",x_awal)
        print("x akhir = ",x_akhir)
        y_awal = float(input("Masukkan nilai y awal = "))
        print("Nilai y awal = ",y_awal)
        print("____________________________________")
        data['Xi'].append(x_awal)
        data['Y'].append(y_awal)
        while True:
            print ("Nilai x =",round((x_awal+h),3))
            k1 = k_1(x_awal,y_awal)
            k2 = k_2(x_awal,y_awal,h,k1)
            k3 = k_3(x_awal,y_awal,h,k1,k2)
            k4 = k_4(x_awal,y_awal,h,k1,k2,k3)
            fungsi_next = y_awal+((1/6)*(k1+2*k2+2*k3+k4)*h)
            print("Nilai y(",round((x_awal+h),3),") = ",fungsi_next)
            x_awal+=h
            y_awal = fungsi_next
            data['Xi'].append(x_awal)
            data['Y'].append(y_awal)
            print("- - - - - - - - - - - - - - -- - - -")
            if x_awal >= x_akhir:
                break
        break
    else:
        print("Pilihan salah")
```
